```python
import numpy as np
import scipy
import scipy.interpolate
import cmath
import matplotlib.pyplot as plt
```

# Tight focusing a LG beam using For loops (possibly to be parallelizible)

```python
# Cartesian to Polar coordinates
def cart2pol(x, y):
    rho = np.sqrt(x**2 + y**2)
    phi = np.arctan2(y, x)
    return(rho, phi)

n=1.0;           # Refractive index of the meadium in which the focusing occurs
NA=0.95;         # Numerical Aperture of the aplanatic lens

N=128;           # Number of points each dimension
k=2*np.pi;       # Wavenumber

xmax=1.5;        # Half lenght of the numerical window
z=0;

k=2*np.pi
xx=np.linspace(-xmax,xmax,N)
X, Y = np.meshgrid(xx, xx)
r,phi= cart2pol(X,Y) # Coordinates de

s=1;
l=1;

THETA=np.linspace(0,np.arcsin(NA/n),2*N)
dth=np.abs(THETA[0]-THETA[2]);
R=np.linspace(0,np.max(r),N)
RR,TT=np.meshgrid(R,THETA)

rp=np.sin(THETA)/(s*np.sin(NA/n))
K= np.sqrt(np.cos(THETA))*np.sin(THETA)*(rp**np.abs(l))*np.exp(-(rp/(np.sqrt(2)))**2)*np.exp(1j*k*z*np.cos(THETA))

def II(RR):
  I1=dth*np.sum(K*(1+np.cos(THETA))*scipy.special.jv(l,k*RR*np.sin(THETA)))
  I2=dth*np.sum(K*(1-np.cos(THETA))*scipy.special.jv(l+2,k*RR*np.sin(THETA)))
  I3=dth*np.sum(K*(1-np.cos(THETA))*scipy.special.jv(l-2,k*RR*np.sin(THETA)))
  I4=dth*np.sum(K*np.sin(THETA)*scipy.special.jv(l+1,k*RR*np.sin(THETA)))
  I5=dth*np.sum(K*np.sin(THETA)*scipy.special.jv(l-1,k*RR*np.sin(THETA)))
  return [I1, I2, I3, I4, I5]


##
II1=np.array([])
II2=np.array([])
II3=np.array([])
II4=np.array([])
II5=np.array([])

for ii in range(N):
  for jj in range(N):
    Int=II(r[jj,ii])
    II1=np.append(II1,Int[0])
    II2=np.append(II2,Int[1])
    II3=np.append(II3,Int[2])
    II4=np.append(II4,Int[3])
    II5=np.append(II5,Int[4])



I1=np.real(np.array(II1).reshape((N,N)))
I2=np.real(np.array(II2).reshape((N,N)))
I3=np.real(np.array(II3).reshape((N,N)))
I4=np.real(np.array(II4).reshape((N,N)))
I5=np.real(np.array(II5).reshape((N,N)))



```

```python
Exx= (1j**l)*np.exp(1j*l*phi)*( I1+0.5*np.exp(1j*2*phi)*I2+0.5*np.exp(-1j*2*phi)*I3)
Exy=-0.5*(1j**(l-1))*np.exp(1j*l*phi)*(-np.exp(1j*2*phi)*I2+np.exp(-1j*2*phi)*I3)
Exz=(1j**(l+1))*np.exp(1j*l*phi)*( np.exp(1j*phi)*I4-np.exp(-1j*phi)*I5)
Eyx=Exy;
Eyy=(1j**l)*np.exp(1j*l*phi)*( I1-0.5*np.exp(1j*2*phi)*I2-0.5*np.exp(-1j*2*phi)*I3)
Eyz=(1j**(l))*np.exp(1j*l*phi)*( np.exp(1j*phi)*I4+np.exp(-1j*phi)*I5)


```

```python
pol=np.array([0*np.pi/4,-np.pi/2])
Ex=np.cos(pol[0])*Exx+np.exp(1j*pol[1])*np.sin(pol[0])*Eyx
Ey=np.cos(pol[0])*Exy+np.exp(1j*pol[1])*np.sin(pol[0])*Eyy
Ez=np.cos(pol[0])*Exz+np.exp(1j*pol[1])*np.sin(pol[0])*Eyz

def PlotComponents(Ex,Ey,Ez):
  plt.figure(figsize=(9, 6))
  plt.subplot(231)
  plt.imshow(abs(Ex)**2)
  plt.axis('off')
  plt.subplot(232)
  plt.imshow(abs(Ey)**2)
  plt.axis('off')
  plt.subplot(233)
  plt.imshow(abs(Ez)**2)
  plt.axis('off')
  plt.subplot(234)
  plt.imshow(np.angle(Ex),cmap='hsv')
  plt.axis('off')
  plt.subplot(235)
  plt.imshow(np.angle(Ey),cmap='hsv')
  plt.axis('off')
  plt.subplot(236)
  plt.imshow(np.angle(Ez),cmap='hsv')
  plt.axis('off')
  plt.colorbar
  plt.show()


# check Calculation

PlotComponents(Ex,Ey,Ez)
```
